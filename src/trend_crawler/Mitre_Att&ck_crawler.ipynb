```python
import requests
from bs4 import BeautifulSoup

mitre_url = "https://attack.mitre.org/"
response = requests.get(mitre_url)
bs4_html = BeautifulSoup(response.text, "html.parser")
bs4_html
```

```python
tactic_list = [tactic.text.strip() for tactic in bs4_html.find("thead").find_all("td" ,{"class" : "tactic name"})] 

technique_info_list = []
techniques_table_list = bs4_html.find_all("table", {"class" : "techniques-table"})
for index, techniques_table in enumerate(techniques_table_list):
    for info in techniques_table.find_all("td", {"class" : "technique"}) : 

        technique_info = {"tactic" : None, "technique" : None, "url" : None}    
        technique_info["tactic"] = tactic_list[index]
        technique_info["technique"] = info.find("a").text.strip()
        technique_info["url"] = "https://attack.mitre.org/" + info.find("a")["href"]
        technique_info_list.append(technique_info)

import pandas
technique_info_df = pandas.DataFrame(technique_info_list)

import re
def technique(technique) : 
    try : 
        pattern = r"\(\d{1,2}\)"
        match = re.search(pattern, technique)
        translate_technique = technique.replace(match.group(), "").replace("\u00a0", "")
        # if "\u00a0" in translate_technique : 
        #     print(translate_technique)
        return translate_technique
    except : 
        return technique

technique_info_df["technique"] = technique_info_df["technique"].map(technique)
technique_info_df.to_json("data/mitre_att&ck.json")
```
