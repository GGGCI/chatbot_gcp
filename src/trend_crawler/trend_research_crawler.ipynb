```python
def parser_meta_from_bs4(post_info, bs4_html) :     
    fields_text = [
        ("title", "h1", {"class" : "article-details__title"}),
        ("post_time", "time", {"class" : "article-details__date"}),
    ]
    for field, element_tuple, attributes in fields_text :
        try : 
            post_info[field] = bs4_html.find(element_tuple, attributes).text
        except : 
            pass

def parser_content_from_bs4(post_info, bs4_html) : 
    try : 
        post_info["content"] = [rtchText.text for rtchText in bs4_html.find_all("div", {"class" : "richText"})]
    except : 
        pass
    
    def parser_TID(post_info, input_string) : 
        pattern = r'T\d{4}(.\d{3}){0,1}'
        matches = [i.group() for i in re.finditer(pattern, input_string)]
        post_info["TID"] = matches

    for index, richText in enumerate(post_info["content"]):
        if "MITRE ATT&CK Matrix" in richText:
            try : 
                parser_TID(post_info, post_info["content"][index+1])
                break
            except : 
                break
    
    post_info["content"] = ". ".join([richText.replace("\n", "") for richText in post_info["content"]])

# post_info_url
def parser_url(post_title) : 
    with open("trendmicro_open_source_reports.txt", encoding="utf-8") as file:
        url_lines = file.read().splitlines()   
    for url in url_lines : 
        if post_title in url : 
            return url
        
import os
from bs4 import BeautifulSoup
import pandas
from tqdm import tqdm
import re

post_info_list = []
total_files = sum(len(files) for _, _, files in os.walk("html"))
for root, dirs, files in os.walk("html") : 
    for file in tqdm(files, total=total_files, desc="Processing Files"):
        with open(os.path.join(root, file), 'r', encoding='utf-8') as post_file:
            bs4_html = BeautifulSoup(post_file, 'html.parser')
        
        post_info = {'title': None, "url" : None, 'post_time': None, "content" : None}
        post_info["url"] = parser_url(file)
        parser_meta_from_bs4(post_info, bs4_html)
        parser_content_from_bs4(post_info, bs4_html)
        post_info_list.append(post_info)

post_info_df = pandas.DataFrame(post_info_list)
def post_time_for_day(post_time) : 
    try : 
        post_day = post_time.split(",")[0].split(" ")[-1].strip().replace("0", "")
        return post_day 
    except : 
        return None

def post_time_for_mouth(post_time) : 
    try : 
        post_mouth = post_time.split(",")[0].split(" ")[0].strip()
        return post_mouth 
    except : 
        return None
    
def post_time_for_year(post_time) : 
    try : 
        post_year = post_time.split(",")[-1].strip()
        return post_year
    except : 
        return None
    
post_info_df["post_day"] = post_info_df["post_time"].map(post_time_for_day)
post_info_df["post_mouth"] = post_info_df["post_time"].map(post_time_for_mouth)
post_info_df["post_year"] = post_info_df["post_time"].map(post_time_for_year)
post_info_df.to_json("data/cybersecurity_intelligence.json")
post_info_df[~(post_info_df["TID"].isna())].copy().to_json("data/TID_fine_tune.json")
```
